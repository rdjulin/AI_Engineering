## A. Build a baseline model (5 marks)



### 1.1 Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_split helper function from Scikit-learn.  

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense

#load dataset
dataset = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
dataset.head()
```

```python
X = dataset.iloc[:,:8]
y = dataset.iloc[:, 8:9]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
n_cols = X_train.shape[1]
```

### 1.2 Train the model on the training data using 50 epochs.

```python
def regr_model():
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,) ))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(1))
    model.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics=['accuracy'])
    return model
```

```python
model = regr_model()

# fit the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
```

### 1.3 Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.



```python
from sklearn.metrics import mean_squared_error
y_pred = model.predict(X_test)

print('Mean Squared Error: '+ str(mean_squared_error(y_pred, y_test)))
```

### 1.4 Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

```python
from sklearn.metrics import mean_squared_error
mse_list = []

for i in range(50):
    #load dataset
    X = dataset.iloc[:,:8]
    y = dataset.iloc[:, 8:9]

    dataset = pd.read_csv('concrete_data.csv')
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
    n_cols = X_train.shape[1]

    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (n_cols,) ))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(1))
    model.compile(optimizer = 'adam', loss = 'mean_squared_error',metrics=['accuracy'])

    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)

    
    y_pred = model.predict(X_test)

    mse_list.append(mean_squared_error(y_pred, y_test))
```

### 1. 5 Report the mean and the standard deviation of the mean squared errors.



```python
import statistics
#standart devision
std_of_mse = statistics.stdev(mse_list)

#mean
summ = sum(mse_list)
countt = len(mse_list)
mean_of_mse = summ/countt

print('The standard devision of the Mean Squared Errors is '+ str(std_of_mse))
print('The mean of the Mean Squared Errors is '+ str(mean_of_mse))
```
